```python
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load ClinicalBERT
tokenizer = AutoTokenizer.from_pretrained("medicalai/ClinicalBERT")
model = AutoModel.from_pretrained("medicalai/ClinicalBERT")

def get_embedding(text):
    """Encodes text into a ClinicalBERT embedding"""
    tokens = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        output = model(**tokens)
    return output.last_hidden_state.mean(dim=1).numpy()  # Take the mean of token embeddings
```

```python
# Sample EHR text (long clinical record)
ehr_text = """The patient exhibits high blood pressure and glucose levels. Previous medical history includes Type 2 Diabetes and hypertension.
They were prescribed Metformin and Lisinopril for management. Recent tests indicate kidney function is deteriorating."""

# Medical condition labels with descriptions
labels = {
    "Diabetes": "A chronic disease characterized by high blood sugar levels.",
    "Hypertension": "A condition in which blood pressure is consistently too high.",
    "Kidney Disease": "Damage or decreased function of the kidneys over time.",
    "Heart Failure": "A chronic condition where the heart does not pump blood effectively.",
    "Obesity": "A condition characterized by excessive body fat accumulation."
}

# Encode EHR and label descriptions into ClinicalBERT embeddings
ehr_embedding = get_embedding(ehr_text)
label_embeddings = {label: get_embedding(desc) for label, desc in labels.items()}

# Compute similarity scores
scores = {label: cosine_similarity(ehr_embedding, label_embedding)[0][0] for label, label_embedding in label_embeddings.items()}

# Normalize scores using softmax to get confidence percentages
exp_scores = np.exp(list(scores.values()))
confidence_scores = {label: (exp_score / sum(exp_scores)) * 100 for label, exp_score in zip(scores.keys(), exp_scores)}

# Print results
for label, score in sorted(confidence_scores.items(), key=lambda x: x[1], reverse=True):
    print(f"✅ {label}: {score:.2f}% confidence")
```

```python
def get_embedding(model_name, text):
    tokenizer = tokenizers[model_name]
    model = models[model_name]

    tokens = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        output = model(**tokens)

    return output.last_hidden_state.mean(dim=1)  # Sentence-level embedding
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Store results
results = {}

for model_name in models:
    results[model_name] = {}

    for relation, pairs in relationship_tests.items():
        similarity_scores = []

        for concept1, concept2 in pairs:
            emb1 = get_embedding(model_name, concept1).numpy()
            emb2 = get_embedding(model_name, concept2).numpy()

            score = cosine_similarity(emb1, emb2)[0][0]
            similarity_scores.append((concept1, concept2, score))

        results[model_name][relation] = similarity_scores

# Create detailed pandas DataFrame with a column for each model
df_data = {}
for model_name in models:
    df_data[model_name] = {}
    for relation, examples in results[model_name].items():
        for concept1, concept2, score in examples:
            key = (relation, concept1, concept2)
            if key not in df_data:
                df_data[key] = {}
            df_data[key][model_name] = score

# Convert to DataFrame
df = pd.DataFrame.from_dict(df_data, orient='index').reset_index()
df.columns = ["Relationship Type", "Concept 1", "Concept 2"] + list(models.keys())

# Save DataFrame for analysis
df.to_csv("model_similarity_details.csv", index=False)

# Generate separate plots for each model
for model_name in results:
    plt.figure(figsize=(10, 5))
    for relation, scores in results[model_name].items():
        sns.kdeplot([s[2] for s in scores], label=f"{relation}")
    plt.xlabel("Cosine Similarity")
    plt.ylabel("Density")
    plt.title(f"Similarity Score Distribution - {model_name}")
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Normalize similarity scores per model using Min-Max Normalization
def min_max_normalize(scores):
    min_val = np.min(scores)
    max_val = np.max(scores)
    return [(s - min_val) / (max_val - min_val) if max_val != min_val else 0.5 for s in scores]

normalized_results = {}
for model_name, relations in results.items():
    normalized_results[model_name] = {}
    all_scores = np.concatenate([scores for scores in relations.values()])
    min_val, max_val = np.min(all_scores), np.max(all_scores)

    for relation, scores in relations.items():
        normalized_results[model_name][relation] = min_max_normalize(scores)

# Visualization: Normalized Similarity Score Distributions
plt.figure(figsize=(14, 7))
for model_name in normalized_results:
    for relation, scores in normalized_results[model_name].items():
        sns.kdeplot(scores, label=f"{model_name} - {relation}")
plt.xlabel("Normalized Cosine Similarity")
plt.ylabel("Density")
plt.title("Normalized Similarity Score Distribution Across Relationship Types")
plt.legend()
plt.show()

# Print aggregated normalized results
for model_name in normalized_results:
    print(f"\nNormalized Results for {model_name}:")
    for relation, scores in normalized_results[model_name].items():
        print(f"  {relation}: Avg Similarity = {np.mean(scores):.2f}, Min = {np.min(scores):.2f}, Max = {np.max(scores):.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Standardize similarity scores per model using Z-Score Standardization
def z_score_standardize(scores):
    mean_val = np.mean(scores)
    std_val = np.std(scores)
    return [(s - mean_val) / std_val if std_val != 0 else 0 for s in scores]

standardized_results = {}
for model_name, relations in results.items():
    standardized_results[model_name] = {}
    all_scores = np.concatenate([scores for scores in relations.values()])
    mean_val, std_val = np.mean(all_scores), np.std(all_scores)

    for relation, scores in relations.items():
        standardized_results[model_name][relation] = z_score_standardize(scores)

# Visualization: Standardized Similarity Score Distributions
plt.figure(figsize=(14, 7))
for model_name in standardized_results:
    for relation, scores in standardized_results[model_name].items():
        sns.kdeplot(scores, label=f"{model_name} - {relation}")
plt.xlabel("Standardized Cosine Similarity (Z-Score)")
plt.ylabel("Density")
plt.title("Standardized Similarity Score Distribution Across Relationship Types")
plt.legend()
plt.show()

# Print aggregated standardized results
for model_name in standardized_results:
    print(f"\nStandardized Results for {model_name}:")
    for relation, scores in standardized_results[model_name].items():
        print(f"  {relation}: Avg Similarity = {np.mean(scores):.2f}, Min = {np.min(scores):.2f}, Max = {np.max(scores):.2f}")
```

```python
snomed_labels = [
    "Chronic lung disease [disorder]",
    "Asthma [disorder]",
    "Chronic obstructive lung disease [disorder]"
]

snomed_embeddings = {
    model_name: torch.vstack([get_embedding(model_name, label) for label in snomed_labels])
    for model_name in models
}

for model_name, emb in snomed_embeddings.items():
    print(f"SNOMED Embeddings ({model_name}): Shape {emb.shape}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_results = {}

for model_name in models:
    ehr_embedding_np = ehr_embeddings[model_name].numpy()
    snomed_embeddings_np = snomed_embeddings[model_name].numpy()

    similarity_scores = cosine_similarity(ehr_embedding_np, snomed_embeddings_np)
    best_match_idx = similarity_scores.argmax()
    best_match_label = snomed_labels[best_match_idx]

    similarity_results[model_name] = {
        "scores": similarity_scores,
        "best_match": best_match_label
    }

    print(f"Model: {model_name} → Best Match: {best_match_label}")
```

```python
threshold = 0.8

for model_name in models:
    similarity_scores = similarity_results[model_name]["scores"]
    matched_labels = [snomed_labels[i] for i in range(len(snomed_labels)) if similarity_scores[0, i] > threshold]

    print(f"Model: {model_name} → Matched Labels: {matched_labels}")
```
